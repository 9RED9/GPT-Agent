### рҹ’» кө¬кёҖ мҪ”лһ©(Google Colab) мӢӨн–үмҡ© мҪ”л“ң

мҪ”лһ© нҷҳкІҪм—җм„ң мғҲлЎңмҡҙ м…Җмқ„ м—ҙкі  м•„лһҳ мҪ”л“ңлҘј к·ёлҢҖлЎң ліөмӮ¬н•ҳм—¬ л¶ҷм—¬л„Јкё° н•ң л’Ө мӢӨн–үн•ҳмӢңл©ҙ лҗ©лӢҲлӢӨ.

```
# 1. н•„мҲҳ лқјмқҙлёҢлҹ¬лҰ¬ м„Өм№ҳ (мөңмҙҲ мӢӨн–ү мӢң м•Ҫк°„мқҳ мӢңк°„мқҙ мҶҢмҡ”лҗ  мҲҳ мһҲмҠөлӢҲлӢӨ)
!pip install -qU deepagents langchain-core langchain-openai
```

```


# 2. нҷҳкІҪ ліҖмҲҳ м„Өм • (OpenAI API нӮӨ мһ…л Ҙ)
import os
import getpass

# API нӮӨк°Җ нҷҳкІҪліҖмҲҳм—җ м—ҶлӢӨл©ҙ м•Ҳм „н•ҳкІҢ мһ…л Ҙл°ӣлҸ„лЎқ мІҳлҰ¬н•©лӢҲлӢӨ.
if "OPENAI_API_KEY" not in os.environ:
    os.environ["OPENAI_API_KEY"] = getpass.getpass("OpenAI API KeyлҘј мһ…л Ҙн•ҳм„ёмҡ”: ")

# 3. м—җмқҙм „нҠё лӘЁл“Ҳ мһ„нҸ¬нҠё л°Ҹ мҪ”л“ң мӢӨн–ү
from deepagents import create_deep_agent
from langchain_core.messages import HumanMessage

EXAMPLE_QUERY = "How do I stream intermediate tool results from a subagent?"

# кё°мӨҖ м—җмқҙм „нҠё(baseline_agent) мҙҲкё°нҷ”
baseline_agent = create_deep_agent(
    model="openai:gpt-5.5",
    tools=[],
    system_prompt=(
        "You are a helpful LangChain documentation assistant. "
        "Answer questions about LangChain APIs and patterns."
    ),
)

# м—җмқҙм „нҠё мӢӨн–үн•ҳм—¬ кІ°кіј л°ӣм•„мҳӨкё°
result = baseline_agent.invoke(
    {"messages": [HumanMessage(content=EXAMPLE_QUERY)]}
)

# кІ°кіј н…ҚмҠӨнҠё м¶ңл Ҙ (мҳӨлҘҳ л°©м§ҖлҘј мң„н•ҙ .text лҢҖмӢ  .content мӮ¬мҡ©)
print(result["messages"][-1].content)
```

**рҹ’Ў мӢӨн–ү нҢҒ:**

* мҪ”л“ң мӢӨн–ү мӢң мһ…л Ҙм°Ҫмқҙ лӮҳнғҖлӮҳл©ҙ мӨҖл№„н•ҳмӢ  `sk-...`лЎң мӢңмһ‘н•ҳлҠ” **OpenAI API Key**лҘј л¶ҷм—¬л„Јкі  м—”н„°лҘј лҲ„лҘҙмӢңл©ҙ лҗ©лӢҲлӢӨ. (мһ…л Ҙ мӢң ліҙм•Ҳмқ„ мң„н•ҙ нҷ”л©ҙм—җ л¬ёмһҗк°Җ н‘ңмӢңлҗҳм§Җ м•ҠмҠөлӢҲлӢӨ.)
* л§Ңм•Ҫ `gpt-5.5` лӘЁлҚём—җ лҢҖн•ң м ‘к·ј к¶Ңн•ңмқҙ м—ҶлӢӨл©ҙ, `model="openai:gpt-4o"` л“ұмңјлЎң лӘЁлҚёлӘ…мқ„ ліҖкІҪн•ҳм—¬ н…ҢмҠӨнҠён•ҙ ліҙмӢңлҠ” кІғмқ„ к¶ҢмһҘн•©лӢҲлӢӨ.
